```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df_values = pd.read_csv('Files/2.5.Histogram-exercise.csv')
df_values
```

```python
freq = df_values.value_counts().sort_index()
freq
```

```python
INTERVAL = 10 #Elijo la cantidad de intervalos que deseo
interval_width = (max(df_values["Data set"]) - min(df_values["Data set"])) / INTERVAL #Calculo la amplitud del intervalo

def all_intervals(interval, interval_width):
    """Devuelve listas con todos los intervalos de inicio y todos los intervalos finales"""

    interval_start = min(df_values["Data set"])
    interval_end = max(df_values["Data set"])
    list_interval_start = []
    list_interval_end = []
    current_interval = interval_start

    for i in range(INTERVAL):
        list_interval_start.append(current_interval)
        current_interval += interval_width
        list_interval_end.append(current_interval)

    return(list_interval_start, list_interval_end)
```

```python
df_intervals = pd.DataFrame()
df_intervals["Interval start"] = all_intervals(INTERVAL, interval_width)[0]
df_intervals["Interval end"] = all_intervals(INTERVAL, interval_width)[1]
df_intervals
```

```python
def absolutes_frecuencys():
    count = 0
    y = 0
    absolute_frecuency = []
    for i in range(INTERVAL): 
        while df_intervals["Interval start"][i] <= df_values["Data set"][y] and df_intervals["Interval end"][i] >= df_values["Data set"][y]:
            print(i,df_intervals["Interval start"][i],df_values["Data set"][y],df_intervals["Interval end"][i])
            count += 1 
            y += 1
        absolute_frecuency.append(count)
        count = 0
    return absolute_frecuency

df_intervals["Absolute frecuency"] = absolutes_frecuencys()
```

```python
df_intervals
```

```python
sum_absolute_frecuency = sum(df_intervals["Absolute frecuency"]) + 1
def relatives_frecuencys():
    relative_frecuency = []
    for i in range(INTERVAL):
        relative_frecuency.append(df_intervals["Absolute frecuency"][i] / sum_absolute_frecuency)
    return relative_frecuency

df_intervals["Relative frecuency"] = relatives_frecuencys()
df_intervals
```
