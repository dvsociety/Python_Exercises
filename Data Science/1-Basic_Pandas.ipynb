```python
#Creando dataframe desde un archivo csv
df_exams = pd.read_csv('StudentsPerformance.csv')
df_exams
```

```python
#Algunas funciones basicas con pandas 
len(df_exams)
max(df_exams.index) #Sin agregar el index estaria devolviendo el maximo de un string
type(df_exams)
round(df_exams, 2) #Redondeo de decimales
```

```python
#Seleccionar columnas de un dataframe
df_exams['gender']
#Mismo manejo de atributos y metodos en la seleccion de columnas 
type(df_exams['gender'])
df_exams['gender'].index
df_exams['gender'].tail()
```

```python
#Agregar una nueva columna al dataframe 
df_exams['language score'] = 70
df_exams
```

```python
#Crear una nueva columna con numpy y sin numeros contantes 
language_score = np.arange(0,1000)
df_exams['language score'] = language_score
df_exams
```

```python
#Columna no numeros enteros aleatorios 
language_score = np.random.randint(1, 100, size=1000)
df_exams['language score'] = language_score
df_exams
```

```python
#Creacion de numeros aleatorios usando decimales 
np.random.uniform(1, 100, size=1000)
```

```python
#Operaciones en columnas 
df_exams['math score'].sum()
df_exams['math score'].count()
df_exams['math score'].mean()
df_exams['math score'].std()
df_exams['math score'].min()
df_exams.describe()
```

```python
#Operaciones en filas
df_exams['average'] = (df_exams['math score'] + df_exams['reading score'] + df_exams['writing score']) / 3
df_exams.round(1)
```

```python
#Contar elementos por categoria 
df_exams['gender'].value_counts()
df_exams['gender'].value_counts(normalize=True) #Obtiene la frecuencia relativa
```

```python
df_exams['parental level of education'].value_counts()
```

```python
#Ordenar un dataframe
df_exams.sort_values(by='math score')
df_exams.sort_values(by='math score', ascending=False) #Ordena en modo descendente 
df_exams.sort_values(by=['math score', 'reading score'], ascending=False) #Ordena tomando como prioridad math score

#sort.values no actualiza los valores a no ser que se lo indique con el argumento inplace=True
# df_exams.sort_values(by='math score', inplace=True)

#Ordenar usando el argumento key y una funcion lambda
df_exams.sort_values(by='race/ethnicity', ascending=True, key=lambda col:col.str.lower()) 
```
